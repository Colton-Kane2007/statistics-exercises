```python
# Do manual cars get better gas mileage than automatic cars?
def is_manual(trans):
    return 'man' in trans
df['is_manuals'] = df['trans'].apply(is_manual)
man= df[df.is_manuals==True].hwy
auto = df[df.is_manuals==False].hwy
t, p = stats.ttest_ind(man, auto)
t, p
```

```python
if p/2 > alpha:
    print("We fail to reject the null hypothesis.")
elif t < 0:
    print("We fail to reject null hypothesis.")
else:
    print("We reject the null hypothesis.")
```

```python
# Use the following contingency table to help answer the question of whether using a Macbook and being a Codeup student are independent of each other.

 	#Codeup Student	Not Codeup Student
#Uses a Macbook	49	20
#Doesn't Use A Macbook	1	30
df1 = pd.DataFrame(
    {
        'codeup_student': [49,1],
        'not_student': [20,30]
    }, index=['macbook_user', 'not_macbook']
)

#State your null and alternative hypotheses.
#null: there is not an association between being a codeup student and using a macbook
#alt: there is an association between being a codeup student and using a macbook
#State your alpha.
alpha = .05
#crosstab
#observed = pd.crosstab(df1.codeup_student, df1.not_student)
#Perform a chi2 test of independence.
chi2, p, degf, expected = stats.chi2_contingency(df1)
chi2, p, degf, expected
#State your conclusion
# p is less than alpha, so we can reject our null hypothesis, there is an association between being a codeup student and using a macbook
```

```python
#Is an employee's gender independent of whether an employee works in sales or marketing? (only look at current employees)

#State your null and alternative hypotheses.
#null: there is an association between gender and the two departments
#alt: there is not an association between gender and the two departments
#State your alpha.
alpha = .05
#crosstab
observed = pd.crosstab(df.gender, df.dept_name)
#Perform a chi2 test of independence.
chi2, p, degf, expected = stats.chi2_contingency(observed)
chi2, p, degf, expected
#State your conclusion
# p is greater than alpha, so we cannot reject our null hypothesis, there is not an association between gender and the two departments
```
